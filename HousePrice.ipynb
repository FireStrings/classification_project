```python
import pandas as pd
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# np.random.seed(5)

dados = pd.read_csv("data/house_price/data.csv")
dados.head()
# a_renomear = {"expected_hours":"horas_esperadas", "price":"preco", "unfinished":"nao_finalizado"}
# dados.rename(columns=a_renomear, inplace=True)

# trocar = {0:1, 1:0}

# dados["finalizado"] = dados.nao_finalizado.map(trocar)

# sns.scatterplot(x="condition", y="price", data=dados)
dados["price"] = dados["price"].apply(int)
x = dados[["sqft_above","sqft_basement","sqft_living"]]
y = dados[["price"]]


treino_x, teste_x, treino_y, teste_y = train_test_split(x,y, test_size=0.25, random_state = 20)


modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
# for i in treino_x.sqft_living:
#     if not isinstance(i,int):
#         print(i)
previsoes = modelo.predict(teste_x)

acuracia = accuracy_score(teste_y, previsoes)*100
previsoes
print("A acurácia foi de : %.2f%% " % acuracia )
```
